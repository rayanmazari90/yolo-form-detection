```python
import os
import random
import shutil

# set the directory where the JPEG and CSV files are located
directory = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/image"

# set the directory for the training and validation sets
training_directory = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/images/training"
validation_directory = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/images/validation"
training_labels_directory = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/labels/training"
validation_labels_directory = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/labels/validation"

# set the split ratio for the training and validation sets
split_ratio = 0.8

# initialize the counters
num_training_files = 0
num_validation_files = 0


# iterate over the files in the directory
for filename in os.listdir(directory):
    # check if the file is a JPEG file
    if filename.endswith(".jpg") or filename.endswith(".jpeg"):
        # get the corresponding CSV filename
        csv_filename = os.path.splitext(filename)[0] + ".csv"
        # randomly assign the file to the training or validation set
        if random.uniform(0, 1) < split_ratio:
            # move the JPEG file to the training directory
            shutil.copy(os.path.join(directory, filename), os.path.join(training_directory, filename))
            # move the CSV file to the training labels directory
            shutil.copy(os.path.join(directory, csv_filename), os.path.join(training_labels_directory, csv_filename))
            num_training_files += 1
        else:
            # move the JPEG file to the validation directory
            shutil.copy(os.path.join(directory, filename), os.path.join(validation_directory, filename))
            # move the CSV file to the validation labels directory
            shutil.copy(os.path.join(directory, csv_filename), os.path.join(validation_labels_directory, csv_filename))
            num_validation_files += 1

print(f"{num_training_files} files were moved to the training set, and {num_validation_files} files were moved to the validation set.")
```

# Data preparation

```python
# create a dictionary with the content we interrested in :
# Load CSV file
df = pd.read_csv('/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/labels/training/Fuente0parte_amistoso_6_26.csv')

df= df[df["campo"]!="texto maquina"]

def clean_no_si(df):
    for elem in df["campo"].unique():
        if "_si" in elem:
            new_elem = elem.replace("_si", "")
            df.loc[df['campo'] == elem, 'campo'] = new_elem
        elif " si" in elem:
            new_elem = elem.replace(" si", "")
            df.loc[df['campo'] == elem, 'campo'] = new_elem
        elif "_no" in elem:
            new_elem = elem.replace("_no", "")
            df.loc[df['campo'] == elem, 'campo'] = new_elem
        elif " no" in elem:
            new_elem = elem.replace(" no", "")
            df.loc[df['campo'] == elem, 'campo'] = new_elem
    return df

df = clean_no_si(df)
encoded_vars = {}
counter =0
for elem in df["campo"].unique():
    encoded_vars[elem]= counter 
    counter +=1
```

Create Labeled Text Data 

```python
from PIL import Image

# set the path to the image file
image_path = "/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/images/training/Fuente0parte_amistoso_6_27.jpg"

# open the image and get its size
with Image.open(image_path) as img:
    width, height = img.size
    print(f"The image size is {width} x {height}")


# Training 
folder_path = '/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/csvfiles/training'  # Path to the test folder

for file_name in os.listdir(folder_path):
        df = pd.read_csv(folder_path+"/"+file_name)
        df= df[df["campo"]!="texto maquina"]
        df = clean_no_si(df)
        file_labeled= open(f"/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/labels/training/{os.path.splitext(file_name)[0]}.txt","w+")
        for index, row in df.iterrows():
            x,y, w, h , elem= row["x"], row["y"], row["w"], row["h"], row["campo"]
            file_labeled.write(f"{encoded_vars[elem]} {(x+ (w//2))/width} {(y+ (h//2))/height} {w/width} {h/height} \n")

# Validation
folder_path ='/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/csvfiles/validation'

for file_name in os.listdir(folder_path):
        df = pd.read_csv(folder_path+"/"+file_name)
        df= df[df["campo"]!="texto maquina"]
        df = clean_no_si(df)
        file_labeled= open(f"/Users/rayanmazari/Desktop/Statistical learning/Group Project/Yolo/yolov5_ws/data/labels/validation/{os.path.splitext(file_name)[0]}.txt","w+")
        for index, row in df.iterrows():
            x,y, w, h , elem= row["x"], row["y"], row["w"], row["h"], row["campo"]
            file_labeled.write(f"{encoded_vars[elem]} {(x+ (w//2))/width} {(y+ (h//2))/height} {w/width} {h/height} \n")
```
